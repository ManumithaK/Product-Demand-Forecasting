```python
import numpy as np
import pandas as pd


from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import train_test_split
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import LabelEncoder, StandardScaler

```

```python
df = pd.read_csv("/Users/manumitha/Desktop/UNT/ML/train.csv")

```

```python
# filling empty rows under 'Item_Weight' column with their mean value
df['Item_Weight'].fillna(df['Item_Weight'].mean(), inplace=True)
```

```python
# filling empty rows under 'Outlet_Size' column with their mean value
df['Outlet_Size'].fillna(df['Outlet_Size'].mode(), inplace=True)
```

```python
df.head()
```

```python
en=LabelEncoder()
df['Item_Identifier']=en.fit_transform(df['Item_Identifier'])
df['Item_Type']=en.fit_transform(df['Item_Type'])
df['Outlet_Identifier']=en.fit_transform(df['Outlet_Identifier'])

```

```python
df = pd.get_dummies(df, columns = ['Item_Fat_Content', 'Outlet_Size', 'Outlet_Location_Type', 'Outlet_Type'])
print(df)
```

```python
X = df.drop(columns=[ 'Item_Outlet_Sales']) 
y = df['Item_Outlet_Sales']
```

```python
# Linear Regression model as Baseline model
# Evaluation metrices used are RMSE, R Square
X_train, X_valid, y_train, y_valid= train_test_split(X,y,test_size=0.3, random_state=42) 
model1=LinearRegression()
model1.fit(X_train, y_train)
y_pred=model1.predict(X_valid) 
rmse=np.sqrt(mean_squared_error(y_pred,y_valid)) 
r2=r2_score(y_pred,y_valid)
print("RMSE Score : ", rmse)
print("R2 Square score : ", r2)
```
